# Generate 3 digit code

 We could have seen 10 digits code in supermarkets that starts from 00000000001. Ever wondered how would you generate these code in a single line? You will not have this in your mind once you read through this code!
 
 
 Lets try adding id to a string

```python
product_id = ['p_'+ str(no) for no in range(1,151)]
product_id[::8]
```

   Though the numbers are generated as expected, while sorting we may have a different order! :(

```python
sorted(product_id)
```

   This is why we need a perfect 3 digit or may be 10 digit. This can be done simply with 'zfill' method! 

```python
product_id = ['p_'+ str(no).zfill(3) for no in range(1,151)]
```

```python
product_id[::8]
```

Lets sort this

```python
sorted(product_id)
```
