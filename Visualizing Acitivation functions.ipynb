# Activation functions

      Artificial Neural Network is a network of Artificial neurons or perceptrons. Each perceptron does two activities. First it calculates a linear combination of inputs and second it calculates the activation of the first. Second part is the activation function and it decides whether or not the output of the node will be considered in the next layer. This allows the neural network to learn more non-linear and complex functions.
    
      These functions can be linear or non-linear depending on the task it performs. This article will be a visualization of some activation functions.

## Linear activation function

    This is the simplest activation function. There is no transformation at all! They are used in regression problem.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,2000)

def lin_act(x):
    return 2*x
y = lin_act(x)

plt.plot(x,y)
plt.title("Linear Activation Function")
plt.ylabel("2 x")
plt.show()
```

## Sigmoid Function
    Sigmoid is a S shaped function which is close to a step function. There output is between 0 and 1 and so they are widely used in probabilistic estimations.  A combination of sigmoid functions called softmax is used for multi class classifications. The function is differentiable and monotonic. 

```python
def sig_fn(x):
    return 1.0/(1+np.exp(-x))

y = sig_fn(x)

def plot_sigmoid(x,y,title="Sigmoid Activation Function"):
    plt.plot(x,y)
    plt.legend(["Sigmoid"])
    plt.title(title)
    plt.hlines(1, -5, 5, linestyle='dashed', color= 'red')
    plt.hlines(0.5, -5, 5, linestyle='dashed', color= 'red')
    plt.hlines(0, -5, 5, linestyle='dashed', color= 'red')
    plt.ylabel("sigmoid")
    plt.show()

    
plot_sigmoid(x,y)
```

    Sensitivity of Sigmoid function can be increased by magnifying the input. (Note: sentivity can only be altered near the y-axis only and far from y axis, it is either 0 or 1.

```python
y = sig_fn(5*x)
plot_sigmoid(x,y,"Sigmoid Activation function for 5x")
```

```python
y = sig_fn(x/3)
plot_sigmoid(x,y,"Sigmoid Activation function for x/3")
```

Derivative of a sigmoid function $\sigma(z)$ is  
$\sigma(z)*(1-\sigma(z))$

# tanh
    Tanh can measure the negative effects as well. Its range is from -1.0 to 1.0.  The function is differentiable and is monotonic. 

```python
y = np.tanh(x)

def plot_tanh(x,y,title="tanh Activation Function"):
    plt.plot(x,y)
    plt.legend(["tanh"])
    plt.title(title)
    plt.hlines(1, -5, 5, linestyle='dashed', color= 'red')
    plt.hlines(0, -5, 5, linestyle='dashed', color= 'red')
    plt.hlines(-1, -5, 5, linestyle='dashed', color= 'red')
    plt.ylabel("tanh")
    plt.show()

plot_tanh(x,y)
```

```python
plt.plot(x,sig_fn(x))
plt.plot(x,np.tanh(x))
plt.hlines(1, -5, 5, linestyle='dashed', color= 'red')
plt.hlines(0, -5, 5, linestyle='dashed', color= 'red')
plt.hlines(0.5, -5, 5, linestyle='dashed', color= 'red')
plt.hlines(-1, -5, 5, linestyle='dashed', color= 'red')
plt.legend(['sigmoid','tanh'])
plt.show()
```

Deriavative of $tanh(x)$ is $1-tanh^{2}(x)$

    A common problem with sigmoid function and tanh is that they saturate for a large numbers. However large a number will be snapped to one. And however small it may be will be snapped to zero. They are sensitive around mid point. They are more significant in deep neural network causing vanishing gradient problem and prevents deep networks from learning more.
    This problem was solved by Relu.

# Relu
    Rectified Linear Unit (Relu) is a piecewise linear function. Which returns the input if it is positive else it returns zero. Both the function and its deriavative are monotonic. It is widely used function in Deep learning.

```python
def Relu(x):
    return [max(i,0) for i in x]

y = Relu(x)

plt.plot(x,y)
plt.title("Rectified Linear Unit")
plt.ylabel("Relu")
plt.show()
```

Its Derivative is 

    1 if x>0,
    0 if x<=0.
             
Variants of Relu:
   1. Leaky Relu
   2. Exponential Linear Unit or ELU
   3. Parametric Relu

Other Activation functions:
