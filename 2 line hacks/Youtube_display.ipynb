# Watch Videos in the terminal

   Bored of opening a new tab or sharing a link for code description? You can have a player playing Youtube video in your kernel. You can play the video from where you got the code reference...
   
 Just import one function YouTubeVideo from Ipython.display and pass the youtube video id (11 digit id).

```python
from IPython.display import YouTubeVideo
start = 1
YouTubeVideo('7UJ4CFRGd-U',start=start,autoplay=True,width=800, height=500)
```

There are more options for customization to resize, snap a particular location and much more.. 

Explore more from https://ipython.readthedocs.io/en/stable/api/generated/IPython.display.html#IPython.display.YouTubeVideo
